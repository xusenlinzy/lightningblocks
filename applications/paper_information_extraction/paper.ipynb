```python
def reformat(input_file):
    import re

    with open(input_file, "r", encoding="utf-8") as f:
        text = f.read()
        texts = text.strip().split("\n")

        results = []
        sentences = []
        special_column = None
        for i, line in enumerate(texts):
            if "专栏" in line:
                if special_column is not None:
                    results.append((special_column, "".join(sentences)))

                special_column = line.strip()
                sentences = []
            else:
                sentences.append(line.strip())
                if i == len(texts) - 1:
                    results.append((special_column, "".join(sentences)))

        data = []
        for res in results:
            sentence_list = re.split(u"[0-9]+\.", res[1])
            sentence_list = list(filter(None, sentence_list))
            for sent in sentence_list:
                sent_list = sent.split("。")
                data.append((res[0], sent_list[0], "".join(sent_list[1:])))

    return data
```

```python
data = reformat("data/专栏.txt")
data
```

```python
import pandas as pd
data = pd.DataFrame(data, columns=["专栏名", "标题", "文本内容"])
data
```

```python
data.to_csv("data/专栏.csv", index=False)
```

```python
import requests

schema = ["专有名词", "疾病", "症状、临床表现", "身体物质和身体部位", "药物", "检查、治疗或预防程序", "部门科室",
          "医学检验项目", "微生物", "检查设备和治疗设备"]

def predict(text, schema):
    data_bin = {
    'model_name': 'uie-paper',
    'input': text,
    "uie_schema": " ".join(schema),
    "device": "gpu",
    }
    res = requests.post('http://192.168.0.55:8000/uie', json=data_bin).json()
    res = list(set([v["text"] for _, vs in res["rlt"].items() for v in vs]))
    return res
```

```python
data["实体"] = data["文本内容"].apply(lambda x: predict(x, schema=schema))
```
